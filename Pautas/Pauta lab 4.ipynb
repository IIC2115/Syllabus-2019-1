# Problema 1

```python
def suma_cuadrados(L):
    L = list(set(L))         
    S2 = {l**2 for l in L}
    L2 = list(S2)    
    R = []       
    for i in range(len(L)):
        x = L2[i]
        for j in range(i,len(L)):
            index = x + L2[j]            
            if (index in S2):                                
                #R.append((L[i], L[j], int(math.sqrt(index))))
                R.append([int(math.sqrt(x)), int(math.sqrt(L2[j])), int(math.sqrt(index))])
    return R

def suma_cuadrados2(L):
    L = list(set(L)) 
    S2 = {x**2 for x in L}    
    return [(int(math.sqrt(x)),int(math.sqrt(y)), int(math.sqrt(x+y))) for x,y in itertools.combinations(S2,2) if (x+y in S2)]
```

# Problema 2

```python
def gremlins(entrada):

    n = len(entrada)
    seConviertePor = [None for x in range(n)]
    diaDemonio = [0 for x in range(n)]
    minimo = entrada[0]
    respuesta = 0
    
    for g in range(n):
        if entrada[g] <= minimo:
            minimo = entrada[g]
        else:
            saltoDeDias = 0
            k = g-1
            while k >= 0:
                if diaDemonio[k] != 0 and saltoDeDias >= diaDemonio[k]:
                    k -= 1
                    continue
                elif entrada[g] > entrada[k]:
                    break
                else:
                    saltoDeDias = max(saltoDeDias, diaDemonio[k])
                    k = seConviertePor[k]
            seConviertePor[g] = k
            diaDemonio[g] = saltoDeDias + 1
            respuesta = max(respuesta, diaDemonio[g])
    return respuesta+1
```

# Problema 3

```python
def luces(arr):
    n = 0
    n1 = 0
    n2 = 0
    if len(arr) == 2:
        if arr[0]==1 or arr[1]==1:
            return 1
    maxPos = arr.index(max(arr))
    sub1 = arr[:maxPos+1]
    sub2 = arr[maxPos:]
    if len(sub1[:-1])>1:
        n1 = luces(sub1[:-1])
    if len(sub2[1:])>1:
        n2 = luces(sub2[1:])
    sub1.sort()
    sub2.sort()
    for eni,i in enumerate(sub1):
        for enj,j in enumerate(sub2):
            if (eni!=len(sub1)-1) or (enj!=len(sub1)-1):
                if i*j<=arr[maxPos]:
                    n+=1
                else:
                    break
    return n1+n2+n
```

Ajustar manualmente `size` de tal manera que el tiempo promedio de ejecutar `suma_cuadrados(L)` sea lo más cercano posible a un segundo.

```python
total_time = 0
runs = 10
size = 8000
for k in range(runs):
    #L = random.sample(range(1, size+1), size)
    L = [int(random.uniform(1,min((10000,size))+1)) for _ in range(size)]
    start = time.time()
    L2 = suma_cuadrados(L)
    end = time.time()
    total_time += end-start 
print(total_time/runs)
```
